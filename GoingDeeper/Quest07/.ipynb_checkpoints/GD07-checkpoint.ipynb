# NSMC 데이터 분석 및 Huggingface dataset 구성

```python
import pandas as pd
from datasets import Dataset

# --- 파일 이름 정의 ---
train_file = "ratings_train.txt"
test_file = "ratings_test.txt"

# --- Pandas로 TSV(Tab-separated values) 파일 읽기 ---
# read_csv를 사용하되, 탭으로 분리되어 있으므로 sep='\t' 옵션을 줍니다.
train_df = pd.read_csv(train_file, sep='\t')
test_df = pd.read_csv(test_file, sep='\t')

# --- 누락된 값(NaN)이 포함된 행 제거 ---
# NSMC 데이터에는 비어있는 리뷰가 일부 포함되어 있어 오류를 유발할 수 있습니다.
train_df = train_df.dropna()
test_df = test_df.dropna()

# --- Pandas DataFrame을 Hugging Face Dataset으로 변환 ---
nsmc_train_dataset = Dataset.from_pandas(train_df)
nsmc_test_dataset = Dataset.from_pandas(test_df)

# --- 결과 확인 ---
print("✅ 로컬 파일 로딩 및 변환 완료!")
print(nsmc_train_dataset)
print(nsmc_test_dataset)
```

* Data description  
id각 파일은 , document, 세 개의 열로 구성됩니다.label  
id: 네이버에서 제공하는 리뷰 아이디  
document: 실제 리뷰  
label: 리뷰의 감정 클래스입니다. (0: 부정적, 1: 긍정적)  
열은 탭으로 구분됩니다(즉, .tsv형식이지만 파일 확장자는 .txt초보자도 쉽게 액세스할 수 있도록 되어 있습니다)  
총 20만 개의 리뷰  
    ratings.txt: 총 20만 개의 리뷰  
    ratings_test.txt: 테스트를 위해 50,000개의 리뷰가 보류됨  
    ratings_train.txt: 훈련에 대한 15만 개의 리뷰  
  
* Characteristics  
모든 리뷰는 140자 미만입니다.  
각 감정 클래스는 동등하게 샘플링됩니다(즉, 무작위 추측은 50% 정확도를 낳습니다)  
10만 개의 부정적인 리뷰(원래는 1~ 4등급 리뷰)  
10만 개의 긍정적 리뷰(원래는 9- 10점 평가 리뷰)  
중립적인 리뷰(원래 평점 5~8점 리뷰)는 제외됩니다.  

# klue/bert-base model 및 tokenizer 불러오기

```python
import transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification

huggingface_tokenizer = AutoTokenizer.from_pretrained('klue/bert-base')
huggingface_model = AutoModelForSequenceClassification.from_pretrained('klue/bert-base')
```

```python
# transform 함수 만들기 

def transform(data):
    return huggingface_tokenizer(
        data['document'],
        truncation = True,
        padding = 'max_length',
        return_token_type_ids = False,
        )
```

```python
# (앞부분의 데이터 로딩 코드는 동일하게 실행되었다고 가정합니다)
# huggingface_tokenizer가 미리 정의되어 있어야 합니다.
# from transformers import AutoTokenizer
# huggingface_tokenizer = AutoTokenizer.from_pretrained("klue/bert-base") # 예시

# --- 1. 학습 데이터셋 토크나이징 ---
# 정의된 transform 함수와 map을 사용하여 전체 학습 데이터를 토큰화합니다.
tokenized_train_dataset = nsmc_train_dataset.map(
    transform, 
    batched=True,
    remove_columns=['id', 'document'] # 토큰화 후 원본 id와 document 컬럼은 제거
)

# --- 2. 테스트 데이터셋 토크나이징 ---
# 전체 테스트 데이터도 동일하게 토큰화합니다.
tokenized_test_dataset = nsmc_test_dataset.map(
    transform, 
    batched=True,
    remove_columns=['id', 'document']
)


# --- 3. 테스트 데이터셋을 검증/테스트용으로 분리 ---
# train_test_split을 사용하여 토큰화된 테스트셋을 50% 비율로 나눕니다.
# 결과는 'train'과 'test' 키를 가진 딕셔너리 형태로 저장됩니다.
val_test_split = tokenized_test_dataset.train_test_split(test_size=0.5)


# --- 4. 최종 데이터셋 변수 할당 ---
# 가독성을 위해 최종 변수들에 각각 할당합니다.
final_train_dataset = tokenized_train_dataset
final_val_dataset = val_test_split['train']    # 분리된 결과의 'train' 부분을 검증용으로 사용
final_test_dataset = val_test_split['test']     # 분리된 결과의 'test' 부분을 최종 테스트용으로 사용


# --- 최종 결과 확인 ---
print("✅ 토크나이징 및 데이터셋 분리 완료!")
print("\n--- 최종 데이터셋 구성 ---")
print(f"학습 데이터셋 (Train): {len(final_train_dataset)}개")
print(f"검증 데이터셋 (Validation): {len(final_val_dataset)}개")
print(f"테스트 데이터셋 (Test): {len(final_test_dataset)}개")

print("\n--- 학습 데이터셋 정보 ---")
print(final_train_dataset)
```

# model 학습 진행해 보기

```python
# --- 라이브러리 호출 ---

import os
import numpy as np
import evaluate
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments, # 훈련 규칙을 정의
    Trainer,            # 훈련을 실행하는 주체
    EarlyStoppingCallback # 조기 종
)

# PEFT 라이브러리에서 LoRA 관련 클래스를 가져옵니다.
from peft import LoraConfig, get_peft_model
```

```python
# --- PEFT (LoRA) 설정 및 모델 적용 ---
# LoRA를 적용할 위치와 파라미터 설정
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["query", "value"],
    lora_dropout=0.05,
    bias="none",
    task_type="SEQ_CLS" # 시퀀스 분류 작업 명시
)

# 기존 모델을 LoRA 모델로 감싸줍니다.
peft_model = get_peft_model(huggingface_model, lora_config)
# 훈련 가능한 파라미터 수 확인
peft_model.print_trainable_parameters()
# ---------------------------------------------
```

```python
# --- 평가 메트릭 함수 정의 ---
combined_metrics = evaluate.combine([
    'accuracy', 
    'f1', 
    'precision', 
    'recall',  
    'matthews_correlation'
])

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return combined_metrics.compute(predictions=predictions, references=labels)


record_steps=1000

# --- 3. 훈련 규칙(Arguments) 설정 ---
# 훈련 과정의 세부 옵션을 설정합니다. (결과 저장 폴더, 평가 시점, 배치 사이즈 등)
training_args = TrainingArguments(

    output_dir='./results',          # 훈련 결과(모델 등)가 저장될 폴더
    num_train_epochs=6,              # 총 훈련 에폭 수
    
    per_device_train_batch_size=8,  # 훈련 시 배치 사이즈
    gradient_accumulation_steps=4,   # 4번의 미니 배치마다 그래디언트 업데이트 (실질 배치 사이즈: 8*4=32)
    per_device_eval_batch_size=32,   # 평가 시 배치 사이즈
    fp16=True,                              # (유지) 메모리 절약 및 속도 향상을 위해 필수
    learning_rate = 2e-5,                         #learning_rate
    weight_decay = 0.01,                        # weight decay
    # --- 이 부분을 추가하거나 수정합니다 ---
    logging_strategy="steps",  # epoch마다 로그를 기록합니다.
    logging_steps=record_steps,           # 1000 스텝마다 로그를 기록
    logging_dir='./logs',      # 로그 파일이 저장될 폴더를 지정합니다.
    # ------------------------------------
    # --- 추천 하이퍼파라미터 추가 ---
    lr_scheduler_type='linear',      # 선형 학습률 스케줄러 사용
    warmup_steps=500,                # 훈련 초기 500 step 동안 학습률을 서서히 증가
    
    eval_strategy="steps",     # 매 epoch이 끝날 때마다 검증 데이터로 평가
    eval_steps=record_steps,
    save_strategy="steps",
    save_steps=record_steps,
    load_best_model_at_end=True,     # 훈련 종료 시 최적의 모델을 로드
    metric_for_best_model='f1',      # 최적 모델을 판단할 기준 (f1 score)
    group_by_length=True, #버케팅(Bucketing)
)

# --- 4. Trainer 객체 생성 ---
# 준비된 모든 구성 요소(모델, 규칙, 데이터, 함수)를 Trainer에 전달합니다.
early_stopping = EarlyStoppingCallback(early_stopping_patience=4)

trainer = Trainer(
    model=peft_model,  #huggingface_model,
    args=training_args,
    # train_dataset=quick_train_dataset,
    # eval_dataset=quick_val_dataset,
    train_dataset=final_train_dataset,
    eval_dataset=final_val_dataset,
    compute_metrics=compute_metrics,
    tokenizer=huggingface_tokenizer, # 패딩(padding)을 동적으로 처리하기 위해 추가
    callbacks=[early_stopping] #  조기 종료
)

# --- 5. 훈련 시작 ---
# 이 한 줄의 코드로 전체 훈련 과정이 실행됩니다.
trainer.train()
```

# 시각화

```python
import matplotlib.pyplot as plt

# 1. 훈련이 완료된 trainer에서 학습 기록을 가져옵니다.
history = trainer.state.log_history

# 2. 모든 Loss와 평가 메트릭을 분리하여 저장합니다.
train_loss = [x['loss'] for x in history if 'loss' in x]
eval_loss = [x['eval_loss'] for x in history if 'eval_loss' in x]
epochs = [x['epoch'] for x in history if 'eval_loss' in x]

# --- 이 부분을 추가합니다 ---
eval_f1 = [x['eval_f1'] for x in history if 'eval_f1' in x]
eval_accuracy = [x['eval_accuracy'] for x in history if 'eval_accuracy' in x]
eval_precision = [x['eval_precision'] for x in history if 'eval_precision' in x]
eval_recall = [x['eval_recall'] for x in history if 'eval_recall' in x]
eval_mcc = [x['eval_matthews_correlation'] for x in history if 'eval_matthews_correlation' in x]
# -------------------------

# 3. Matplotlib으로 Loss 커브 시각화
plt.figure(figsize=(15, 6)) # 그래프 전체 크기 조정

plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label='Training Loss', marker='o')
plt.plot(epochs, eval_loss, label='Validation Loss', marker='o')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True) # 그리드 추가

# 4. Matplotlib으로 모든 성능 지표 커브 시각화
# --- 이 부분을 수정합니다 ---
plt.subplot(1, 2, 2)
plt.plot(epochs, eval_f1, label='F1 Score', marker='o')
plt.plot(epochs, eval_accuracy, label='Accuracy', marker='o')
plt.plot(epochs, eval_precision, label='Precision', marker='o')
plt.plot(epochs, eval_recall, label='Recall', marker='o')
plt.plot(epochs, eval_mcc, label='MCC', marker='o')
plt.title('Validation Performance Metrics')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
plt.grid(True) # 그리드 추가
# -------------------------

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn.functional as F

def predict_sentiment(text):
    """
    입력된 텍스트의 긍정/부정 감성을 예측하고 확률을 반환하는 함수
    """
    # 1. 토크나이저로 텍스트를 모델 입력 형태로 변환
    #    - return_tensors="pt": PyTorch 텐서로 결과를 반환하도록 지정
    inputs = huggingface_tokenizer(
        text, 
        return_tensors="pt",
        truncation=True,
        padding=True
    )

    # 2. 모든 입력 데이터를 모델과 동일한 장치(GPU 또는 CPU)로 이동
    inputs = {key: val.to(trainer.model.device) for key, val in inputs.items()}

    # 3. 그래디언트 계산을 비활성화하여 추론 속도 향상
    with torch.no_grad():
        # 모델에 입력을 전달하여 로짓(logits)을 얻음
        logits = trainer.model(**inputs).logits

    # 4. 로짓에 소프트맥스 함수를 적용하여 확률값으로 변환
    probabilities = F.softmax(logits, dim=1).squeeze()

    # 5. 가장 높은 확률을 가진 클래스와 그 확률값을 찾음
    predicted_class_id = torch.argmax(probabilities).item()
    confidence = probabilities[predicted_class_id].item()
    
    # 클래스 ID를 레이블 이름으로 변환
    labels = ["부정", "긍정"]
    predicted_label = labels[predicted_class_id]

    # 결과 출력
    print(f"입력 문장: \"{text}\"")
    print(f"▶️ 예측 결과: {predicted_label} (확신도: {confidence:.2%})")
    print("-" * 50)

# --- 테스트 해볼 예제 문장들 ---
example_reviews = [
    "와 이 영화 진짜 대박! 시간 가는 줄 몰랐어요.",
    "돈 아깝다... 내 인생 최악의 영화.",
    "배우들 연기는 좋은데 스토리가 너무 아쉽네요.",
    "그냥 저냥 볼만 했어요. 킬링타임용.",
    "이걸 영화라고 만들었냐? 감독 나와라",
    "보는 내내 눈물을 멈출 수가 없었습니다... 인생 영화 등극!"
]

# --- 각 문장에 대해 예측 실행 ---
for review in example_reviews:
    predict_sentiment(review)
```
