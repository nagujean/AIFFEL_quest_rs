```python
# 파일 불러오기
# 'Avengers.txt' 파일이 코랩 환경에 업로드되어 있어야 합니다.
with open('/content/sample_data/06TheAvengers.txt', 'r', encoding='utf-8') as f:
    raw_text = f.read()

print("--- 원본 텍스트 ---")
print(raw_text[:500]) # 텍스트 앞부분 500자만 출력하여 확인
```

```python
import re

# 모든 문자를 소문자로 변환
cleaned_text = raw_text.lower()
print("\n--- 소문자 변환 후 (일부) ---")
print(cleaned_text[:500])

# 모든 특수기호를 제거 (알파벳, 숫자, 공백 제외)
# re.sub(패턴, 대체문자, 원본문자열)
# [^a-z0-9\s] : 알파벳 소문자(a-z), 숫자(0-9), 공백(\s)이 아닌 모든 문자
cleaned_text = re.sub(r'[^a-z0-9\s]', '', cleaned_text)
print("\n--- 특수기호 제거 후 (일부) ---")
print(cleaned_text[:500])
```

```python
# 단어 단위로 분리(띄어쓰기 기준)

#split()메서드는 여러 개의 공배도 하나로 처리하여 빈 문자열 토큰을 생성하지 않음

tokens = cleaned_text.split()
print("\n---토큰화 후 (일부)---")
print(tokens[:20])
```

```python
# 딕셔너리 생성
 # 단어별 빈도수를 딕셔너리 형태로 저장

from collections import Counter

# 가장 간단한 방법 .. 다시 딕셔너리로 만들어야함
#word_counts = Counter(tokens)
#dic = dict(word_counts)


# for 문을 사용하는 가장 일반적 방법

dic = {}

for i in tokens:
  if i in dic:
    dic[i] += 1
  else:
    dic[i] = 1



#확인하고 싶다면?....(전체는 너무 길어지니까 한번 정도만 확인하자 )
print(type(dic))
print(dic)
for word, count in word_counts.most_common(10):
    print(f"{word}: {count}")
```

```python
# input()으로 소문자 텍스트를 입력받으면 정수를 return하는 encoder 함수

def encoder(text_input):
    """
    주어진 텍스트를 전처리하고 토큰화한 후, Vocab을 사용하여 정수 시퀀스로 변환합니다.
    Vocab에 없는 단어는 UNK 토큰의 인덱스로 변환합니다.

    Args:
        text_input (str): 인코딩할 소문자 텍스트 문자열.

    Returns:
        list: 정수 인덱스로 변환된 시퀀스.
    """
    # 텍스트 전처리 (소문자 변환 및 특수기호 제거)
    # 입력이 이미 소문자라고 가정하지만, 혹시 모를 경우를 대비하여 다시 수행합니다.
    processed_text = text_input.lower()
    processed_text = re.sub(r'[^a-z0-9\s]', '', processed_text)

    # 토큰화 (띄어쓰기 기준)
    input_tokens = processed_text.split()

    # 정수 시퀀스로 변환
    encoded_sequence = []
    for token in input_tokens:
        # vocab 딕셔너리에 토큰이 있으면 해당 인덱스 사용
        # 없으면 UNK 토큰의 인덱스 사용 (get 메서드 활용)
        encoded_sequence.append(vocab.get(token, vocab['UNK']))

    return encoded_sequence

# --- encoder 함수 테스트 ---
print("\n--- encoder 함수 테스트 ---")

test_sentence_1 = "i am a boy"
encoded_1 = encoder(test_sentence_1)
print(f"입력: '{test_sentence_1}'")
print(f"출력: {encoded_1}")

test_sentence_2 = "obama is a boy'"
encoded_2 = encoder(test_sentence_2)
print(f"\n입력: '{test_sentence_2}'")
print(f"출력: {encoded_2}")
# 'obama'가 vocab에 없다면, 해당 위치에 UNK 토큰의 인덱스가 들어가야 합니다.


```
